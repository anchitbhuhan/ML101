```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from mpl_toolkits import mplot3d
```

```python
data = pd.read_csv("parkinsons_updrs.data.csv")
```

```python
X1 = data['Jitter:PPQ5']
X2 = data['Jitter:DDP']
X3 = data['Jitter:RAP']
X4 = data['Shimmer']
Y = data['total_UPDRS']
X1 = np.array((X1 - X1.min())-(X1.max() - X1.min()))
X2 = np.array((X2 - X2.min())-(X2.max() - X2.min()))
X3 = np.array((X3 - X3.min())-(X3.max() - X3.min()))
X4 = np.array((X4 - X4.min())-(X4.max() - X4.min()))
Y = np.array((Y - Y.min())-(Y.max() - Y.min()))
```

```python
x1_train, x1_test, x2_train, x2_test,x3_train, x3_test,x4_train, x4_test, y_train, y_test = train_test_split(X1, X2,X3,X4, Y, test_size=0.2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x1_train, x2_train, x3_train,x4_train, y_train, 'g.')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x1_test, x2_test,x3_test,x4_test, y_test, 'g.')
```

```python
def hypothesis(a,b,c,d,e,x1,x2,x3,x4):
    return a * x1 + b * x2 + c*x3+d*x4+e
```

```python
def error(a,b,c,d,e,x1,x2,x3,x4,y):
    e = 0
    m = len(x1)
    for i in range(m):
        e += np.power((hypothesis(a,b,c,d,e,x1[i],x2[i],x3[i],x4[i]) - y[i]), 2)
    
    return (1/(2*m)) * e
```

```python
def step_gradient(a,b,c,d,e,x1,x2,x3,x4,y,learning_rate):
    grad_a = 0
    grad_b = 0
    grad_c = 0
    grad_d = 0
    grad_e = 0
    m = len(x1)
    for i in range(m):
        grad_a += 2/m * (hypothesis(a,b,c,d,e,x1[i],x2[i],x3[i],x4[i]) - y[i]) * x1[i]
        grad_b += 2/m * (hypothesis(a,b,c,d,e,x1[i],x2[i],x3[i],x4[i]) - y[i]) * x2[i]
        grad_c += 2/m * (hypothesis(a,b,c,d,e,x1[i],x2[i],x3[i],x4[i]) - y[i]) * x3[i]
        grad_d += 2/m * (hypothesis(a,b,c,d,e,x1[i],x2[i],x3[i],x4[i]) - y[i]) * x4[i]
        grad_e += 2/m * (hypothesis(a,b,c,d,e,x1[i],x2[i],x3[i],x4[i]) - y[i])
        
    a = a - (grad_a * learning_rate)
    b = b - (grad_b * learning_rate)
    c = c - (grad_c * learning_rate)
    d = d - (grad_d * learning_rate)
    e = e - (grad_e * learning_rate)
    
    return a, b, c,d,e
```

```python
def descend(initial_a, initial_b, initial_c,initial_d,initial_e,x1, x2,x3,x4, y, learning_rate, iterations):
    a = initial_a
    b = initial_b
    c = initial_c
    d = initial_d
    e = initial_e
    for i in range(iterations):
        er = error(a, b, c,d,e, x1, x2,x3,x4, y)
        if i % 1000 == 0:
            print(f"Error: {er}, a: {a}, b: {b}, c: {c}, d: {d}, e: {e}")
        
        a, b, c,d,e = step_gradient(a, b, c,d,e ,x1, x2,x3,x4, y, learning_rate)

    return a, b, c,d,e
```

```python
a = 0
b = 1
c = 1
d = 1
e = 1
learning_rate = 0.01
iterations = 10000

final_a, final_b, final_c, final_d, final_e = descend(a, b, c,d,e, x1_train, x2_train,x3_train,x4_train, y_train, learning_rate, iterations)
```
